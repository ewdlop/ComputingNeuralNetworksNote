```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Define the activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, x * alpha)

def softmax(x):
    exp_x = np.exp(x - np.max(x))
    return exp_x / exp_x.sum(axis=0)

# Generate input data
x = np.linspace(-10, 10, 400)

# Calculate the outputs of the activation functions
sigmoid_y = sigmoid(x)
tanh_y = tanh(x)
relu_y = relu(x)
leaky_relu_y = leaky_relu(x)
softmax_y = softmax(np.array([x, x]))

# Plot the activation functions
plt.figure(figsize=(10, 8))

plt.subplot(2, 3, 1)
plt.plot(x, sigmoid_y, label="Sigmoid")
plt.title("Sigmoid")
plt.grid(True)

plt.subplot(2, 3, 2)
plt.plot(x, tanh_y, label="Tanh", color="orange")
plt.title("Tanh")
plt.grid(True)

plt.subplot(2, 3, 3)
plt.plot(x, relu_y, label="ReLU", color="green")
plt.title("ReLU")
plt.grid(True)

plt.subplot(2, 3, 4)
plt.plot(x, leaky_relu_y, label="Leaky ReLU", color="red")
plt.title("Leaky ReLU")
plt.grid(True)

plt.subplot(2, 3, 5)
plt.plot(x, softmax_y[0], label="Softmax", color="purple")
plt.title("Softmax")
plt.grid(True)

plt.tight_layout()
plt.show()
```
